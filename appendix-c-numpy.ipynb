## **Python Lists vs Numpy Arrays**

Key difference and advantages of arrays.
1. lists are inbuilt whereas numpy arrays need to imported
2. Multiple data types can be appended into a list whereas only single data type can be appended in numpy array
3. List can be expanded while numpy array cant expanded
4. Arrays store less memory, and are faster.

## Summarizing operations

Summarizing operations process and array and return a single number 

```python
np.random.seed(2)
pred = np.random.rand(3).round(2)
pred_sum = pred.sum()
pred
```

```python
pred_sum
```

```python
print('min = %.2f' % pred.min())
print('mean = %.2f' % pred.mean())
print('max = %.2f' % pred.max())
print('std = %.2f' % pred.std())
```

For two-dimentional array it works in the same way:

```python
np.random.seed(2)
matrix = np.random.rand(4, 3).round(2)
matrix
```

```python
matrix.max()
```

But we can specify the axis along which we apply the summarizing operation

- `axis=1` - apply to each rows
- `axis=0` - apply to each column

```python
matrix.max(axis=1)
```

```python
matrix.max(axis=0)
```

```python
matrix.sum(axis=1)
```

### Inverse

To inverse a matrix, use `inv` function from `linalg` package

```python
A = np.array([
    [0, 1, 2],
    [1, 2, 3],
    [2, 3, 3]
])
```

```python
Ainv = np.linalg.inv(A)
Ainv
```

```python
A.dot(Ainv)
```

B is not invertible ("singular"):

```python
B = np.array([
    [0, 1, 1],
    [1, 2, 3],
    [2, 3, 5]
])

np.linalg.inv(B)
```

Note: When you need to solve _Ax = b_, you don't really to compute the inverse. You can use `solve`:

```python
b = np.array([1, 2, 3])
```

```python
x = np.linalg.solve(A, b)
x
```

```python
A.dot(x)
```
